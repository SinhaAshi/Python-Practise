```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1,3,10)
y=np.random.randint(1,3,10)

print(x,y,sep='\n')
```

```python
x = np.arange(1,11)
y = np.sin(x)
z = np.cos(x)

print(x,y,z, sep = '\n\n')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, label = 'SINQ')
plt.plot(y,z, label = 'COSQ')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (6,4))
plt.plot(x,y,label = 'SINQ', color = 'red')
plt.plot(y,z, label ='COSQ')
plt.title('SINQ and COSQ')
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-', alpha = 0.3,marker = "*", markerfacecolor='yellow', markersize = 10, markeredgecolor='blue', linewidth = 10)
plt.show()
```

```python
plt.figure(figsize = (6,4))

plt.subplot(1,2,1)
plt.title('SINQ and COSQ')
plt.plot(x,y,label = 'SINQ', color = 'red')
plt.plot(y,z, label ='COSQ')
plt.xlabel("Data Points")
plt.ylabel("Associate values")

plt.subplot(1,2,2)
plt.title('SINQ and COSQ')
plt.plot(x,y,label = 'SINQ', color = 'red')
plt.plot(y,z, label ='COSQ')
plt.xlabel("Data Points2")
plt.ylabel("Associate values2")

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8,4))

plt.subplot(3,3,1)
plt.title('SINQ and COSQ')
plt.plot(x,y,label = 'SINQ', color = 'red')
plt.plot(y,z, label ='COSQ')
plt.xlabel("Data Points")
plt.ylabel("Associate values")

plt.subplot(3,3,5)
plt.title('SINQ and COSQ')
plt.plot(x,y,label = 'SINQ', color = 'red')
plt.plot(y,z, label ='COSQ')
plt.xlabel("Data Points2")
plt.ylabel("Associate values2")

plt.subplot(3,3,9)
plt.title('SINQ and COSQ')
plt.plot(x,y,label = 'SINQ', color = 'red')
plt.plot(y,z, label ='COSQ')
plt.xlabel("Data Points2")
plt.ylabel("Associate values2")
plt.grid()
plt.legend()
plt.show()
```
