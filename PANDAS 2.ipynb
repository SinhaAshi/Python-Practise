```python
import pandas as pd
df = pd.read_csv("tips.csv")
df
```

```python
df['day'].value_counts()
#Saturday is the busiest day
```

```python
df['time'].value_counts()
#PPeople mostly come for dinner
```

```python
df['sex'].value_counts()
#Male customers are more
```

```python
df.groupby('day')['total_bill'].sum()
#Sat is the highest selling day
```

```python
df.groupby('sex')['tip'].mean()
#Male gives more tips
```

```python
df.groupby('day')['total_bill'].mean()
#Average spending everyday
```

```python
df['total_bill'].max()
#Highest tip
```

```python
df['size'].median()
#Typical group size is 2
```

```python
df.nlargest(5, 'tip')
```
