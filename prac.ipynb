```python
class Bank:
    loc = "Haldia"
    IFSC = 1235
    Branch = "main"
    sector = 5
    
    def __init__(self, name, age, gender,balance):
        self.name = name
        self.age = age
        self.gender = gender
        self.balance = balance
    #object method -- to update object members
    def display(self):
        print(self.name, self.age, self.gender, self.balance)
    
    #class method
    @classmethod #without this object member can be modified but no the static members..
    def change_branch(self,new_branch):
        self.Branch=new_branch

obj1 = Bank("Ash",19,"Female",50000)
obj1.display()
obj1.change_branch("localite")
print(obj1.Branch)
obj1.change_branch("server")
print(Bank.Branch)
print(obj1.Branch)
```
