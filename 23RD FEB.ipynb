```python
#using constructor channing method using super()
class ten:
    def __init__(self,name,roll,marks): #constructor for class ten to initialize three var
        self.name=name
        self.roll=roll
        self.marks=marks

    def display(self): #methid of class ten
        print(self.name, self.roll, self.marks)

class twelve(ten): #twelve inherits ten
    def __init__(self,name,roll,marks,stream,phone): #another constructor for class twelves own data
        super(). __init__(name,roll,marks) #to call the parent class constructor
        self.stream=stream #add new properties to class twelve
        self.phone=phone
    def show_details(self):
        super().display() #calls parent class method
        print(self.stream,self.phone)
c1 = twelve("Ashi",123,98,"PCB",9834097644)
c1.show_details()
        




```

```python
#Multilevel inheritance
class Bachelors(twelve):
    def __init__(self,name,roll,marks,stream,phone,degree,year):
        super(). __init__(name,roll,marks,stream,phone)
        self.degree=degree
        self.year=year
    def show(self):
        super().show_details()
        print(self.degree,self.year)
c2 = Bachelors("Ashi",12,98,"PCB",268755,"Btech",2025)
c2.show()
print(Bachelors.mro()) #METHOD RESOLUTION ORDER
```

```python
#Heirarchial inheritance
class A():
    a=10
    b=20
    def __init__(self,c,d):
        self.c=c
        self.d=d
class B(A):
    a=1000
    @classmethod
    def display(cls):
        print(cls.a)
class C(A):
    a=900
obj1=B(100,80)
obj2=C(90,102)
obj1.display()
print(obj2.a)
print(obj1.b)
        
```

```python
class Person():
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def display(self):
        print(self.name,self.age)

class Student(Person):
    def __init__(self,name,age,roll,standard):
        super(). __init__(name,age)
        self.roll=roll
        self.standard=standard
    def display_2(self):
        super().display()
        print(self.roll,self.standard)

class Academics():
    def __init__(self,course=None,cgpa=None):
        self.course=course
        self.cgpa=cgpa

class Sports():
    def __init__(self,sports=None,level=None):
        self.sports=sports
        self.level=level

##

class AllRounder(Student, Academics, Sports):
    def __init__(self, name, age, roll, standard,
                 is_academic=False, is_sports=False,
                 course=None, cgpa=None,
                 sports=None, level=None):

        super().__init__(name, age, roll, standard)

        self.is_academic = is_academic
        self.is_sports = is_sports

        if self.is_academic:
            Academics.__init__(self, course, cgpa)

        if self.is_sports:
            Sports.__init__(self, sports, level)

    def display_all(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Roll: {self.roll}")
        print(f"Standard: {self.standard}")

        if self.is_academic:
            print(f"Course: {self.course}")
            print(f"cgpa: {self.cgpa}")
        if self.is_sports:
            print(f"Sports name:{self.sports}")
            print(f"level:{self.level}")
ars = AllRounder(
    name="Ashi",
    age=20,
    roll=101,
    standard= 12,
    is_academic=True,
    is_sports=True,
    course="PCM",
    cgpa=9.2,
    sports="Badminton",
    level="State"
)

ars2 = AllRounder(
    name="Ashi",
    age=20,
    roll=101,
    standard= 12,
    is_academic=True,
    is_sports=False,
    course="PCM",
    cgpa=9.2
)

ars.display_all()
ars2.display_all()

    
```
